```python
import kagglehub
import os
import pandas as pd
# Download latest version
path = kagglehub.dataset_download("anshulmehtakaggl/60k-responses-of-16-personalities-test-mbt")

df = pd.read_csv(os.path.join(path, os.listdir(path)[0]), encoding='latin1')
df.sample(10)
```

```python
X = df[df.columns[1:-1]]
y = df[df.columns[-1]]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y.to_numpy().reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
score
```

```python
import numpy as np
m = '''
  Strongly Agree    :  2
  Agree             :  1
  Neutral           :  0
  Disagree          : -1
  Strongly Disagree : -2 
'''
print(m)
user_input = [int(input(q)) for q in X.columns]
data = pd.DataFrame([user_input], columns=X.columns)

probs = model.predict(data)

predicted_mbti = encoder.categories_[0][np.argmax(probs)]
print("Predicted MBTI:", predicted_mbti)
```

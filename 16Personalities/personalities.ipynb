### THis is MBTI personality test, using dataset from Kaggle

```python
import kagglehub
import os
import pandas as pd
import numpy as np

path = kagglehub.dataset_download("anshulmehtakaggl/60k-responses-of-16-personalities-test-mbt")

df = pd.read_csv(os.path.join(path, os.listdir(path)[0]), encoding='latin1')
df.sample(13)
```

```python
df.shape
```

##### separate X and y (features and labels)

```python
X = df[df.columns[1:-1]]
y = df[df.columns[-1]]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)

y_labels = encoder.inverse_transform(y)
y_labels
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_test.shape, X_train.shape
```

### Pipeline for seamless flow for end-to-end product

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
classifier = RandomForestClassifier(
  n_estimators=155, 
  random_state=42, 
  max_depth=50, 
  min_samples_split=10, 
  max_features='sqrt'
)

pipeline = make_pipeline(scaler, classifier)

pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

#### accuracy_score vs f1_score for my model

```python
from sklearn.metrics import accuracy_score, f1_score
print("accuracy score: ", accuracy_score(y_test, y_pred))
print("f1 score: ", f1_score(y_test, y_pred, average='macro'))
```
